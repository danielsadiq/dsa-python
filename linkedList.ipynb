```python
class Node:
  def __init__(self, data, next=None):
    self.data = data
    self.next = next

class LinkedList:
  def __init__(self):
    self.head = None
  
  def insert_at_beginning(self, data):
    node = Node(data, self.head)
    self.head = node
    
  def insert_at_end(self, data):
    if self.head is None:
      self.insert_at_beginning(data)
    itr = self.head
    while itr.next:
      # print(itr.data)
      itr = itr.next
    itr.next = Node(data)
    
  def print(self):
    if self.head is None:
      print("Linked List is empty")
      return
    llstr = ""
    itr = self.head
    while itr:
      llstr += f"{str(itr.data)}-> " 
      itr = itr.next
    print(llstr)     
```

```python
ll = LinkedList()
ll.insert_at_beginning(5)
ll.insert_at_beginning(25)
ll.insert_at_end(35)
ll.insert_at_end(55)
ll.print()
```

Arrays are stored contiguosly in memory, i.e. stored together, one by one.

Hence, insert and delete operations at the beginning or the middle are O(n)
Inserting at the end is/can be O(n)
Traversing is O(n)
Indexing is O(1)

Arrays are of two types, static and dynamic arrays.
Static array, the length of the list is predefined. 
In dynamic, it grows by itself.
